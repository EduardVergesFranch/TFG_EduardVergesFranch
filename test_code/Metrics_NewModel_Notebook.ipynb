#### Percentage Miss Detections

```python
percentages = []
titles = []
for an, aud in zip(anno, audio):
    met = NewModel_Metrics(an,aud,info = False)
    p, _ = met.miss_detection_errors()
    titles.append(met.title)
    percentages.append(p)
    
```

#### Pitch Class Set Representation

```python
support_dataset = list(np.zeros(48))

for an, aud in zip(anno, audio):
    met = NewModel_Metrics(an,aud,info = False)
    if desired_notes:
        dn = met.notes_durations[desired_notes]
    support = met.pitch_class_set_support_histogram(hist = False,desired_notes = dn)
    
    support_dataset = [a + b for a, b in zip(support_dataset, support )] ## Add to global result

fig, ax = plt.subplots(figsize = (15,5))
ax.bar(range(len(support_dataset)), support_dataset, align='center')
ax.set_xticks(range(len(support_dataset)))
ax.set_xticklabels(met.list_pitch_classes)

ax.set_ylabel('Freq.')
ax.set_xlabel('Pitch Class Sets')
ax.set_title('Frequency of each Pitch Class Set in Dataset')
fig.tight_layout()
plt.grid(axis='y', alpha=0.75)
plt.xticks(rotation=45)
plt.show()
```

#### Overall Pitch Assessment Accuracy

```python
from IPython.display import display, Markdown
```

```python
overall_dataset_accuracy = []

for an, aud in zip(anno, audio):
    met = NewModel_Metrics(an,aud,model[4],info = False)
    if desired_notes:
        dn = met.notes_durations[desired_notes]
    acc = met.overall_accuracy(desired_notes = dn)
    overall_dataset_accuracy.append(acc)
    
    display(Markdown('**Over. Acc for {}**: {}'.format(an.split('/')[-1].replace('.json',''),acc)))
```

```python
display(Markdown('**Average Over. Acc**: {}'.format(np.round(np.mean(overall_dataset_accuracy),2))))
display(Markdown('**Median Over. Acc**: {}'.format(np.round(np.median(overall_dataset_accuracy),2))))
```

#### Pitch Class Set Performance

```python
overall_precision = np.zeros(48)
overall_recall = np.zeros(48)

for num, (an, aud) in enumerate(zip(anno, audio)):
    met = NewModel_Metrics(an,aud,info = False)
    if desired_notes:
        dn = met.notes_durations[desired_notes]
    precision, recall,_ , _ = met.per_class_performance(desired_notes = dn,print_results = True)
    
    overall_precision = [a + b for a, b in zip(overall_precision, precision )] ## Add to global result
    overall_recall = [a + b for a, b in zip(overall_recall, recall )] ## Add to global result
    
overall_precision = [p/(num+1) for p in overall_precision]
overall_recall = [r/(num+1) for r in overall_recall]
```

```python
fig, ax = plt.subplots(figsize = (15,5))
x = np.arange(len(overall_precision))  # the label locations
width = 0.40  # the width of the bars
    
ax.bar(x - width/2, overall_precision, width, label='Precision',align = 'center')
ax.bar(x+ width/2, overall_recall, width, label='Recall',align = 'center')

present = np.where(np.array(support_dataset) > 0)[0]
for p in present:
    ax.axvline(x=p, color='r', linestyle='--')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_xlabel('Pitch Class Sets')
ax.set_title('Average performance by class')
ax.set_xticks(x)

ax.set_xticklabels(met.list_pitch_classes)
ax.legend()


fig.tight_layout()
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

#### Overall Confusion Matrix

```python
import itertools
```

```python
overall_conf_matrix = np.zeros((48,48))
for num, (an, aud) in enumerate(zip(anno, audio)):
    met = NewModel_Metrics(an,aud,info = False)
    if desired_notes:
        dn = met.notes_durations[desired_notes]
    conf_matrix = met.conf_matrix(desired_notes = dn)
    
    overall_conf_matrix = np.add(overall_conf_matrix, conf_matrix)
    
```

```python
cmap = plt.cm.Blues
title = 'Overall Confusion matrix'

plt.figure(figsize = (15,15))
plt.imshow(overall_conf_matrix, interpolation='nearest', cmap=cmap)
plt.title(title)
plt.colorbar(fraction=0.046, pad=0.04)
tick_marks = np.arange(len(met.list_pitch_classes))

plt.xticks(tick_marks, met.list_pitch_classes, rotation=45)
plt.yticks(tick_marks, met.list_pitch_classes)

fmt = '.0f'
thresh = overall_conf_matrix.max() / 2.

for i, j in itertools.product(range(overall_conf_matrix.shape[0]), range(overall_conf_matrix.shape[1])):
    if overall_conf_matrix[i,j] > 0:
        plt.text(j, i, format(overall_conf_matrix[i, j], fmt),
                 horizontalalignment="center",
                 verticalalignment = 'center',
                 color="white" if overall_conf_matrix[i, j] > thresh else "black")
plt.grid(alpha = 0.3)
plt.tight_layout()
plt.ylabel('True label')
plt.xlabel('Predicted label')
            
```
